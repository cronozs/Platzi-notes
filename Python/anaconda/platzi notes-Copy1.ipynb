```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
economista_url = "https://www.eleconomista.com.mx/"
financiero_url = "https://www.elfinanciero.com.mx/"
miled_url = "https://miled.com/"
categori_links = []
lista_de_categorias = []
news_urls = []
all_news = []
news_links = []
```

```python
miled = requests.get(miled_url)
```

```python
miled_bd = BeautifulSoup(miled.text, 'lxml')
```

```python
categorias = miled_bd.find_all('nav', attrs = {'class':'main-navigation'})
```

```python
categorias = categorias[1]
```

```python
categori_list = categorias.find_all('li')
```

```python
categori_list[1].a.get('href')
```

```python
for categori in categori_list:
    link = categori.a.get('href')
    categori_links.append(link)
    str_cat = str(categori.a.contents)
    str_cat = str_cat.replace("[","")
    str_cat = str_cat.replace("]","")
    str_cat = str_cat.replace("'","")
    lista_de_categorias.append(str_cat)
```

```python
for link in categori_links:
    actual_category = requests.get(link)
    actual_category_bd = BeautifulSoup(actual_category.text, 'lxml')
    news = actual_category_bd.find_all('div', attrs = {'class':'post-thumbnail'})
    all_news = all_news + news
for new in all_news:
    url = new.a.get('href')
    news_links.append(url)
```

```python
len(news_links)
```

```python
for link in news_links:
    scraped_new = requests.get(link)
    scraped_new_bd = BeautifulSoup(scraped_new.text, 'lxml')
    content_new = scraped_new_bd.find('div', attrs ={'class':'s-post-content s-post-small-el bb-mb-el'})
    title_new = scraped_new_bd.find('h1',attrs = {'class':'entry-title s-post-title bb-mb-el'})
    print(title_new.contents)
    print(content_new)
    time.sleep(30)
```
